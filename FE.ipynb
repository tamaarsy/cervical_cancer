```python
from sklearn.base import BaseEstimator, TransformerMixin

class selectionFiture(BaseEstimator, TransformerMixin):
    def __init__(self, drop_columns ):
        self.drop_columns = drop_columns
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X.drop(self.drop_columns, axis=1)
```

```python
attr_df = selectionFiture(drop_columns=efficient_feature)
efficient_sf_dt_drop = attr_df.transform(df)
```

```python
efficient_sf_dt_drop_columns = list(efficient_sf_dt_drop.columns)

for i in efficient_sf_dt_drop_columns:
    efficient_sf_dt_drop.rename(columns={i : str(i)+'efficient'}, inplace=True)
```

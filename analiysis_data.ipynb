```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import cv2 as cv
import scipy as sy
from PIL import Image

import os
import pathlib
import glob
```

```python
train_Dykeratotic = list(pathlib.Path('../data/train/Dyskeratotic/').glob('*bmp'))
train_Koilocytotic = list(pathlib.Path('../data/train/Koilocytotic/').glob('*bmp'))
train_Metaplastic = list(pathlib.Path('../data/train/Metaplastic/').glob('*bmp'))
train_Parabasal = list(pathlib.Path('../data/train/Parabasal/').glob('*bmp'))
train_Superfical = list(pathlib.Path('../data/train/Superficial-Intermediate/').glob('*bmp'))
```

```python
class_plot = {'dykeratotic' : train_Dykeratotic,
              'koilocytotic' : train_Koilocytotic,
              'metaplastic' : train_Metaplastic,
              'paraasal' : train_Parabasal,
              'superfical' : train_Superfical}

for img in class_plot.keys():
    folder_name = class_plot[img]
    fig = plt.figure(figsize=(15, 15))
    columns = 5
    rows = 1
    print(f'class : {img}')
    
    for index in range(1, columns*rows+1):
        fig.add_subplot(rows, columns, index)
        image = Image.open(folder_name[index])
        #print(image)
        plt.imshow(image)
        plt.axis('off')
    plt.show()
```

### Histogram

```python
def rgb_hitogram(file_loc):
    imgread = cv.imread(file_loc)

    color = ('b','g','r')
    for k,color in enumerate(color):
        histogram = cv.calcHist([imgread],[k],None,[256],[0,256])
        plt.plot(histogram,color = color)
        plt.xlim([0,256])
    plt.show()
```

```python
def histogram(file_loc):
    data = cv.imread(file_loc)
    histr = cv.calcHist([data], [0], None, [256], [0,256])
    plt.plot(histr)
    plt.show()
```
